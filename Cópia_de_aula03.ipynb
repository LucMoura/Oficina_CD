>Existem dois tipos príncipais de problemas evolvendo conjuntos de dados, problemas de **Classificação** e probelmas de **Regressão.**

## Classificação $\text{x}$ Regressão

- **Classificação:** Os problemas normalmente envolvem classificar as variaveis em dois grupos, geramente $[0, 1]$. Dessa forma, nosso objetivo é identificar os padrões que envolvem a escolha de determinado grupo.
- **Regressão:** Diferente da classificação, aqui estamos tentando encontrar a melhor relação entre as variaveis para prever um valor **contínuo**, ou seja, um valor indeterminado com intervalos infinitos.

> Com base neste conhecimento, é possivel identificar que nosso **DataFrame de tweets**, estão divididos em  **dois grupos**(Positivos e Negativos), sendo assim, um problema de **Classificação**.

> Já o DataFrame sobre preço de **imóveis**, está relacioando variaveis para determinar um **valor contínuo**(Velor do aluguel), dessa forma identificamos um problema de **Regressão**.

## Imputação

Imputação consiste na técninca de lidar com valores vazios, seja removendo suas linhas, ou as substituindo pela média, moda ou mediana.

> Como vimos anteriormente, ac colunas *floor, animal e furniture* são variaveis do tipo Object, **texto**. Vamos explorar elas com mais detalhes.

```python
# Visualizando os valores unicos em cada coluna
df_casas[['animal', 'furniture', 'floor']].unique()
```

```python
df_casas.isnull().sum()
```

> Verificamos as colunas textuais, pois nelas, valores vazios podem não serem exibidos na função `.isnull()`

- Mesmo com valores vazios na coluna *'floor'*, eles não foram computados por conta de não estarem categorizados como **NAN**.

```python
df_casas['floor'].value_counts()
```

- Para ser possível a imputação, precisamos converter estes valores vazios para NAN.

> Para isso, transformaremos o ' - ' em NAN, tornando possível a imputação de valores vazios pela média ou dropar mais fácil.

```python
import numpy as np
```

```python
df_casas['floor'] = df_casas['floor'].replace('-', np.nan)
```

```python
df_casas['floor'].value_counts()
```

```python
df_casas.isnull().sum()
```

> Agora podemos tomar decisões sobre como tratar estes valores vazios.

- **Média**
  - Desvio Padrão
- **Moda**
- **Mediana**
- **Remoção Total:** Remoção de todos os valores faltantes, importante para colunas inuteis.

```python
df_casas.shape
```

```python
# Decisão de substituir pela média
df_casas.fillna(df_casas['floor'].mean(), inplace=True)
```

```python
# Decisão de substituir pela moda
df_casas.fillna(df_casas['floor'].mode()[0], inplace=True)
```

```python
# Decisão de substituir pela mediana
df_casas.fillna(df_casas['floor'].median(), inplace=True)
```

```python
# Decisão de remover os valores vazios
df_casas.dropna(subset=['floor'], inplace=True)
```

```python
print(df_casas.shape)
df_casas.isnull().sum()
```
